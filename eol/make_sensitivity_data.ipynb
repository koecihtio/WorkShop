```python
import pandas as pd
import numpy as np
import codecs as cd
```

```python
df_list = []
for year in range(2004, 2022):
    path = path = '../../data/eol/sensitivity/sensitivity_' + str(year) + '.csv'
    df = pd.read_csv(path, encoding='utf_8_sig', skiprows=4)
    df = df[['[証券コード(提出者)]', '[企業名(提出者)]', '[決算日]', '[ヒット文書]']]
    df.set_axis(['stock_code', 'firm_name', 'ymd', 'doc_hit'], axis=1, inplace=True)
    df['year'] = df['ymd'].apply(lambda x: str(x)[0:4])
    df['stock_code'] = df['stock_code'].apply(lambda x: str(x) if not str(x) == '--' else np.nan)
    df['sensitivity'] = 1
    df_list.append(df)

df = pd.concat(df_list, axis=0)
df.dropna(subset=['stock_code'], how='any', inplace=True)
```

```python
path = '../../data/eol/sensitivity/sensitivity_all.csv'
df.to_csv(path, encoding='utf_8_sig', header=True, index=False)
```
